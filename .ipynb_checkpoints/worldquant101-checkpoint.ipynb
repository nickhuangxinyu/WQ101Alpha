```python
import pandas as pd
pd.set_option('mode.use_inf_as_na', True)
import os
import sys 
import time
import numpy as np
import matplotlib.pyplot as plt 

num_class = 10
path = '/root/ricequant/'
df = pd.read_csv(path+'merge.csv.gz', dtype={'ticker':str})

def ShowMemory():
    import psutil
    import os
    info = psutil.virtual_memory()
    print u'内存使用：',psutil.Process(os.getpid()).memory_info().rss
    print u'总内存a：',info.total
    print u'内存占比：',info.percent
    print u'cpu个数：',psutil.cpu_count()

ShowMemory()
```

```python
df['dollar_volume'] = df['close'] * df['volume']
df['vwap'] = df['total_turnover']/df['volume']
df['returns'] = df.groupby('ticker')['close'].apply(lambda x: x.diff(1) / x)
cap = df.groupby('date')['total_turnover'].sum()
cap.name = 'cap'
df.merge(cap.to_frame(), on='date', how='outer')
df['yr'] = df.groupby('ticker')['close'].apply(lambda x: x.diff(-1) / x)
df.loc[:, 'cap'] = df.groupby('date')['total_turnover'].sum()
df = df[df['date'] > '2006']

ShowMemory()
```

```python
def AddFactorLabel(df, num_class, factor_name):
    df.loc[:, factor_name+'_label'] = df.groupby('date', as_index=False).apply(lambda x: pd.qcut(x[factor_name], q=num_class, labels=False, duplicates='drop')).reset_index(level=0)[factor_name]

def PlotFactor(df, num_class, factor_name):
    temp = df.groupby(factor_name+'_label')['yr'].mean().to_dict()
    plt.title('%s corr %lf' %(factor_name, df['yr'].corr(df[factor_name])))
    plt.bar(range(len(temp.keys())), [temp[i] for i in temp.keys()])
    plt.show()
    
```

```python
#alpha6 : -corr(open,volume, 10)
df.loc[:, 'alpha6'] = df.groupby('ticker').apply(lambda x : -x['open'].rolling(10).corr(x['volume'].rolling(10))).reset_index(level=0)[0]
AddFactorLabel(df, num_class, 'alpha6')
PlotFactor(df, num_class, 'alpha6')
ShowMemory()
```

```python
#alpha33: rank(-1*(1-open/close))
df.loc[:, 'alpha33'] = df.groupby('ticker').apply(lambda x: (x['open']/x['close']).rank()).reset_index(level=0)[0]
AddFactorLabel(df, num_class, 'alpha33')
PlotFactor(df, num_class, 'alpha33')
ShowMemory()
```

```python
#alpha4: ts_rank(rank(low), 9)
df.loc[:, 'low_rank'] = df.groupby('date')['low'].rank()
df.loc[:, 'alpha4'] = - df.groupby('ticker')['low_rank'].rolling(9).apply(lambda x:np.argsort(x)[0]).reset_index(level=0)['low_rank']
AddFactorLabel(df, num_class, 'alpha4')
PlotFactor(df, num_class, 'alpha4')
ShowMemory()
```

```python
#alpha41: sqrt(high*low)-vwap
df.loc[:, 'alpha41'] = (df['high']*df['low'])**0.5-df['vwap']
AddFactorLabel(df, num_class, 'alpha41')
PlotFactor(df, num_class, 'alpha41')
ShowMemory()
```

```python
#alpha44: -corr(high, rank(volume), 5)
df.loc[:, 'volume_rank'] = df.groupby('date')['volume'].rank()
df.loc[:, 'alpha44'] = -df.groupby('ticker').apply(lambda x: x['high'].rolling(5).corr(x['volume_rank'].rolling(5))).reset_index(level=0)[0]
AddFactorLabel(df, num_class, 'alpha44')
PlotFactor(df, num_class, 'alpha44')
ShowMemory()
```

```python
#alpha54: -((low-close)*(open**5))/((low-high)*(close**5))
df.loc[:, 'alpha54'] = -(df['low'] - df['close']) * (df['open']**5)/((df['low'] - df['high'])*df['close']**5)
AddFactorLabel(df, num_class, 'alpha54')
PlotFactor(df, num_class, 'alpha54')
ShowMemory()
```

```python
# alpha101: (close-open) / (high-low + 0.001)
df.loc[:, 'alpha101'] = (df['close'] - df['open']) / (df['high'] - df['low'] + 0.001)
AddFactorLabel(df, num_class, 'alpha101')
PlotFactor(df, num_class, 'alpha101')
ShowMemory()
```

```python
#alpha3: -corr(rank(open), rank(volume), 10)
df.loc[:, 'open_rank'] = df.groupby('date')['open'].rank()
df.loc[:, 'alpha3'] = -df.groupby('date').apply(lambda x: x['open_rank'].rolling(10).corr(x['volume_rank'].rolling(10))).reset_index(level=0)[0]
AddFactorLabel(df, num_class, 'alpha3')
PlotFactor(df, num_class, 'alpha3')
ShowMemory()
```

```python
#alpha5: -rank(open-mean(vwap, 10))*abs(rank(close-vwap))
df['alpha5'] = df.groupby('ticker').apply(lambda x: (x['open']-x['vwap'].rolling(10).mean()).rank()).reset_index(level=0)[0] * (df['close'] - df['vwap']).rank()
AddFactorLabel(df, num_class, 'alpha5')
PlotFactor(df, num_class, 'alpha5')
ShowMemory()
```

```python
# alpha12: -sign(delta(volume, 1))*delta(close, 1)
df['alpha12'] = -np.sign(df['volume'].diff(1))*(df['close'].diff(1))
AddFactorLabel(df, num_class, 'alpha12')
PlotFactor(df, num_class, 'alpha12')
ShowMemory()
```

```python
# alpha13: -rank(cov(rank(close), rank(volume), 5))
df['close_rank'] = df.groupby('date')['open'].rank()
df['alpha13_mid'] = df['close_rank'].rolling(5).cov(df['volume_rank'].rolling(5))
df['alpha13'] = -df.groupby('date')['alpha13_mid'].rank()
AddFactorLabel(df, num_class, 'alpha13')
PlotFactor(df, num_class, 'alpha13')
ShowMemory()
```

```python
# alpha16: -rank(cov(rank(high), rank(volume), 5))
df['high_rank'] = df.groupby('date')['high'].rank()
df['alpha16_mid'] = df['high_rank'].rolling(5).cov(df['volume_rank'].rolling(5))
df['alpha16'] = -df.groupby('date')['alpha16_mid'].rank()
AddFactorLabel(df, num_class, 'alpha16')
PlotFactor(df, num_class, 'alpha16')
ShowMemory()
```

```python
# alpha23: mean(high, 20) < high ? -delta(high, 2) : 0
df['alpha23'] = np.where(df.groupby('ticker')['high'].rolling(20).mean().reset_index(level=0)['high']- df['high'] < 0, -df.groupby('ticker')['high'].diff(2), 0)
AddFactorLabel(df, num_class, 'alpha23')
PlotFactor(df, num_class, 'alpha23')
ShowMemory()
```

```python
# alpha25: rank(-returns*adv20*vwap*(high-close))
df['adv20'] = (df.groupby('ticker')['dollar_volume'].rolling(20).sum()/df.groupby('ticker')['volume'].rolling(20).sum()).reset_index(level=0)[0]
df['alpha25_mid'] = -df['returns']*df['adv20']*df['vwap']*(df['high']-df['low'])
df['alpha25'] = df.groupby('date')['alpha25_mid'].rank()
AddFactorLabel(df, num_class, 'alpha25')
PlotFactor(df, num_class, 'alpha25')
ShowMemory()
```

```python
# alpha38: -rank(ts_rank(close, 10)) * rank(close/open)
df['alpha38_mid'] = df.groupby('ticker')['close'].rolling(10).apply(lambda x: np.argsort(x)[0]).reset_index(level=0)['close']
df['alpha38'] = df.groupby('date')['alpha38_mid'].rank().reset_index(level=0)['alpha38_mid'] * df.groupby('date').apply(lambda x: np.argsort(x['close']/x['open'])[0])
AddFactorLabel(df, num_class, 'alpha38')
PlotFactor(df, num_class, 'alpha38')
ShowMemory()
```
