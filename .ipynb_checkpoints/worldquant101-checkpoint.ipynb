```python
import pandas as pd
pd.set_option('mode.use_inf_as_na', True)
import os
import sys 
import time
import numpy as np
import matplotlib.pyplot as plt 

num_class = 10
path = '/root/ricequant/'
df = pd.read_csv(path+'merge.csv.gz', dtype={'ticker':str})
```

```python
cap = df.groupby('date')['total_turnover'].sum()
cap.name = 'cap'
df.merge(cap.to_frame(), on='date', how='outer')
df['yr'] = df.groupby('ticker')['close'].apply(lambda x: x.diff(-1) / x)
df['vwap'] = df['total_turnover']/df['volume']
df.loc[:, 'cap'] = df.groupby('date')['total_turnover'].sum()
df = df[df['date'] > '2006']
```

```python
def AddFactorLabel(df, num_class, factor_name):
    df.loc[:, factor_name+'_label'] = df.groupby('date', as_index=False).apply(lambda x: pd.qcut(x[factor_name], q=num_class, labels=False, duplicates='drop')).reset_index(level=0)[factor_name]

def PlotFactor(df, num_class, factor_name):
    temp = df.groupby(factor_name+'_label')['yr'].mean().to_dict()
    plt.title('%s corr %lf' %(factor_name, df['yr'].corr(df[factor_name])))
    plt.bar(range(len(temp.keys())), [temp[i] for i in temp.keys()])
    plt.show()
```

```python
#alpha6 : -corr(open,volume, 10)
df.loc[:, 'alpha6'] = df.groupby('ticker', as_index=False).apply(lambda x : -x['open'].rolling(10).corr(x['volume'].rolling(10))).reset_index(level=0)[0]
AddFactorLabel(df, num_class, 'alpha6')
PlotFactor(df, num_class, 'alpha6')
```

```python
#alpha33: rank(open/close)
df.loc[:, 'alpha33'] = df.groupby('ticker', as_index=False).apply(lambda x: (x['open']/x['close']).rank()).reset_index(level=0)[0]
AddFactorLabel(df, num_class, 'alpha33')
PlotFactor(df, num_class, 'alpha33')
```

```python
#alpha4: ts_rank(rank(low), 9)
df.loc[:, 'low_rank'] = df.groupby('date')['low'].rank()
df.loc[:, 'alpha4'] = - df.groupby('ticker')['low_rank'].rolling(9).apply(lambda x:np.argsort(x)[0]).reset_index(level=0)['low_rank']
AddFactorLabel(df, num_class, 'alpha4')
PlotFactor(df, num_class, 'alpha4')
```

```python
#alpha41: sqrt(high*low)-vwap
df.loc[:, 'alpha41'] = (df['high']*df['low'])**0.5-df['vwap']
AddFactorLabel(df, num_class, 'alpha41')
PlotFactor(df, num_class, 'alpha41')
```

```python
#alpha44: -corr(high, rank(volume), 5)
df.loc[:, 'volume_rank'] = df.groupby('date')['volume'].rank()
df.loc[:, 'alpha44'] = -df.groupby('ticker').apply(lambda x: x['high'].rolling(5).corr(x['volume_rank'].rolling(5))).reset_index(level=0)[0]
AddFactorLabel(df, num_class, 'alpha44')
PlotFactor(df, num_class, 'alpha44')
```

```python
#alpha54: -((low-close)*(open**5))/((low-high)*(close**5))
df.loc[:, 'alpha54'] = -(df['low'] - df['close']) * (df['open']**5)/((df['low'] - df['high'])*df['close']**5)
AddFactorLabel(df, num_class, 'alpha54')
PlotFactor(df, num_class, 'alpha54')
```
